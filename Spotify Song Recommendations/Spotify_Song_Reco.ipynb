```python
from warnings import filterwarnings
filterwarnings("ignore")
import os
os.chdir("E:/Data Science/DataSet/")
import pandas as pd
A = pd.read_csv("spotify.csv")
pd.set_option("display.max_columns",30)
pd.set_option("display.max_rows",200)
```

```python
A.info()
```

```python
for i in A.columns:
    if(A[i].dtype=="object"):
        x = A[i].mode()[0]
        A[i] = A[i].fillna(x)
    else:
        x = A[i].mean()
        A[i] = A[i].fillna(x)
```

```python
Q = A[["track_popularity","duration_ms"]]
from sklearn.preprocessing import StandardScaler
ss = StandardScaler()
Q1 = pd.DataFrame(ss.fit_transform(Q),columns=Q.columns)
```

```python
Q1
```

```python
from sklearn.cluster import KMeans
km = KMeans(n_clusters=4)
model = km.fit(Q1)
model.labels_
```

```python
Q['Cluster']=model.labels_
Q["track_artist"] = A.track_artist
Q["track_name"] = A.track_name
Q["track_popularity"] = A.track_popularity
Q["duration_ms"] = A.duration_ms
```

```python
Q
```

```python
w = input("Enter Music Name")
cls = Q[Q.track_name==w].Cluster.values[0]
T = list(Q[Q.Cluster==cls]["track_name"].values)
T.remove(w)
ls = print(f"Related Song{T}")
```
