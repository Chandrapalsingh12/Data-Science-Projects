# Scrapping Data using Requests & BeautifulSoup

```python
import requests
from bs4 import BeautifulSoup
```

```python
uri = 'https://www.onlinereadfreebooks.com/en/Harry-Potter-and-the-Philosophers-Stone/1'
```

```python
data = requests.get(uri)
```

```python
soup = BeautifulSoup(data.content)
```

```python
text = soup.find_all("div",class_="text-muted pt-3")[0].text
```

# Proceesing of Data

```python
def tex_proc(val):
    from re import sub
    text = sub("[^a-zA-Z0-9 ]","",val)
    return text
```

```python
r = ''
for i in text.split(" "):
    r = r + " "+ tex_proc(i)    
```

```python
l = []
for i in r.split(" "):
    l.append(i.upper())
```

```python
Q = list(filter(None, l))
```

```python
Q
```

# Arrange Data into Input And Output Formats

```python
inp = []
op = []
for i in range(3,len(Q),1):
    w0 = Q[i-3]
    w1 = Q[i-2]
    w2 = Q[i-1]
    inp.append(w0+" "+w1)
    op.append(w2)
    
```

# Create DataFrame of Input and Output

```python
import pandas as pd
df = pd.DataFrame([inp,op]).T
df.columns = ["Input","Output"]
```

```python
df
```

```python
df.shape
```

# Get Unique Values

```python
import numpy as np
```

```python
uniq = np.unique(Q)
```

```python
len(uniq)
```

# Find lookup dictionary for uniques

```python
ul = {}
for i in range(0,len(uniq),1):
    ul[uniq[i]]=i
   
```

```python
ul
```

# Create Cube

```python
no_line = df.shape[0]
no_uniq = len(uniq)
seq = 2
```

```python
df.shape[0]
```

# Using Zeros Metrics Prepare X

```python
Xprov = np.zeros((no_line,no_uniq,seq))
```

```python
Xprov.shape
```

```python
for line_no,inp_seq in enumerate(inp):
    for wor_line,word in enumerate(inp_seq.split(" ")):
        Xprov[line_no,ul[word],wor_line]=1
```

# Using Zeros Metrics Prepare Y

```python
Yprov = np.zeros((no_line,no_uniq))
```

```python
Yprov.shape
```

```python
for line_no,inp_seq in enumerate(op):
    Yprov[line_no,ul[inp_seq]]=1
        
    
```

# Create NN Model 

```python
from keras.models import Sequential
from keras.layers import LSTM,Dense,Dropout
from keras.activations import softmax
```

```python
nn = Sequential()
nn.add(LSTM(60,input_shape=(1163,2)))
nn.add(Dense(400))
nn.add(Dropout(0.2))
nn.add(Dense(1163,activation='softmax'))
nn.compile(loss="categorical_crossentropy",metrics=["accuracy"],optimizer="adam")
model = nn.fit(Xprov,Yprov,validation_split=0.1,epochs=140,batch_size=31)
```
