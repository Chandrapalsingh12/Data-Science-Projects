# Replace Null Values using Mean and Mode

```python
def replace(train):    
    for i in train:
        if(train[i].dtype=="object"):
            x = train[i].mode()[0]
            train[i] = train[i].fillna(x)
        else:
            x = train[i].mean()
            train[i] = train[i].fillna(x)      
```

# ANOVA

```python
def Anova(df,cat,con):
    from statsmodels.formula.api import ols
    from statsmodels.stats.anova import anova_lm
    x = ols(f"{con} ~ {cat}",df).fit()
    x_main = anova_lm(x)
    Q = pd.DataFrame(x_main)
    a = Q['PR(>F)'][cat]
    return round(a,3)
    
        
```

```python
remove_val = []
for i in train.columns:
    if(train[i].dtype =="object"):
        print(f"ANOVA TESTING WITH {i} AND SalePrice",Anova(train,i,"SalePrice"))
        if(Anova(train,i,"SalePrice")>=0.05):
            remove_val.append(i)
        
```
